```python
#import needed modules
from datetime import datetime
import matplotlib
import matplotlib.pyplot as pyplot
import numpy as np
```

```python
#create an array of 200 elements at 1sec intervals 
#this is indices of different time perisod (for example if we want a box at 150s, 150s is a indices)
time_series_index = pd.date_range('1/1/2020', periods=200, freq='s')
print(time_series_index)
```

```python
#Create time series data by assign Random Integer Values
time_series = pd.Series(np.random.randint(0,500, len(time_series_index)), index=time_series_index)

#print out the first few values to understand how the data looks like 
#(using head() functions gives overview of the data).
print(time_series.head())
```

```python
#resample 1 second bin to minute and sim the corresponding values
time_series_min = time_series.resample('1Min').sum()

#print out the first few values to understand how the data looks like 
print(time_series_min.head())
```

```python
#time zone conversion - assume original time series was in UTC and we want US EST

time_series_utc = time_series.tz_localize('UTC')
print(time_series_utc.head())
```

```python
#The Time Values of above are all Zeros as per the Universal Time Code (Greenwich mean time). So when we the above output
#to the "US/Eastern" timezone then it shows, the time is +5hours from the Greenwich mena time.
#Greenwich is in England, thats where the time started and its really a myth.
```

```python
#convert to US EST
time_series_est = time_series_utc.tz_convert('US/Eastern')
print(time_series_est.head())
```
